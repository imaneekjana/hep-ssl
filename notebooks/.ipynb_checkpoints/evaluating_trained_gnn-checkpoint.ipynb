```python
"""

Evaluating the model

"""

parquet_dir = '/global/cfs/cdirs/m4474/aneek/particlemind/scripts/data/p8_ee_tt_ecm365_parquet'

dataset_train = CLDHits(parquet_dir, "train", nfiles=10, by_event=True)
dataset_val = CLDHits(parquet_dir, "val", nfiles=10, by_event=True)


myModel.eval()

myModel.projection_head = nn.Identity()

latents = []
energies = []


for x in dataset_val:

    x_hits = x['calo_hit_features']
    energy = np.sum(x_hits[:,3])
    
    graph = EventGraphBuilder()(x_hits) #This is a graph corresponding to the given event

    z = myModel(graph)

    latents.append(z)
    energies.append(energy)

latents = torch.cat(latents)
```

```python
#


from sklearn.manifold import TSNE

tsne = TSNE(
    n_components=2,
    perplexity=30,
    learning_rate=200,
    n_iter=1000,
    random_state=42
)

from sklearn.preprocessing import StandardScaler

latents_scaled = StandardScaler().fit_transform(latents.detach().numpy())
z_2d = tsne.fit_transform(latents_scaled)

labels_np = np.array(energies)
labels_norm = (labels_np - labels_np.min()) / (labels_np.max() - labels_np.min())

    
```

```python
plt.scatter(
    z_2d[:, 0],
    z_2d[:, 1],
    c=labels_norm,
    cmap="plasma",
    s=10
)
plt.colorbar(label="Normalized Label")
```
